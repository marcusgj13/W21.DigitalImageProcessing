```python
y = 128
x = 128
centY = y/2
centX = x/2
yy,xx = np.indices((y,x))
radialDist = np.sqrt((xx-centX)**2 + (yy - centY)**2)
circ1 = radialDist < 10
circ2 = radialDist < 30


plt.figure(figsize=(15,5))
plt.subplot(131)
plt.gca().set_title('Radial distribution')
plt.imshow(radialDist, cmap='inferno', interpolation='nearest')
plt.subplot(132)
plt.gca().set_title('Circle with radius 10')
plt.imshow(circ1, cmap='inferno', interpolation='nearest')
plt.subplot(133)
plt.gca().set_title('Circle with radius 30')
plt.imshow(circ2, cmap='inferno', interpolation='nearest')


plt.show()
```

```python
def convert2gray(imIn):
    newIm = [] #placeholder for the new image
    newIm = imIn[:,:,0]*0.299 + imIn[:,:,1]*0.587 + imIn[:,:,2]*0.114 
    # The above is the ITU-R BT.601 conversion
    
    return newIm
```

```python
pepGray = convert2gray(mpimg.imread('./peppers.tiff'))
print("The number of dimensions of the image is: ", pepGray.ndim)
print("The size of the image is: ", pepGray.shape)
print("The maximum value is: ", pepGray.max())
print("The minimum value is: ", pepGray.min())
plt.imshow(pepGray,cmap='gray')
plt.show()
```

```python
from skimage.color import rgb2gray

pepGraySk = rgb2gray(mpimg.imread('./peppers.tiff'))
print("The number of dimensions of the image is: ", pepGraySk.ndim)
print("The size of the image is: ", pepGraySk.shape)
print("The maximum value is: ", pepGraySk.max())
print("The minimum value is: ", pepGraySk.min())
plt.imshow(pepGraySk,cmap='gray')
plt.show()
```
